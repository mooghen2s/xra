```python
import edge_tts
import os
import shutil
from IPython.display import Audio, display
def main() -> None:
    for i in range(start_num, end_num):
        os.makedirs('tmp', exist_ok=True)
        di_file = f"tmp/output_{i}.txt"
        OUTPUT_FILE = f"tmp/output_{i}.mp3"
        with open(di_file, 'r', encoding='utf-8') as re_file:
            TEXT = re_file.read()
        VOICE = "zh-CN-YunxiaNeural"
        communicate = edge_tts.Communicate(TEXT, VOICE, rate="-10%")
        with open(OUTPUT_FILE, "wb") as file:
            for chunk in communicate.stream_sync():
                if chunk["type"] == "audio":
                    file.write(chunk["data"])
                elif chunk["type"] == "WordBoundary":
                    continue
        print(f'done fot {di_file}')
    for i in range(start_num, end_num):
        OUTPUT_FILE = f"tmp/output_{i}.mp3"
        dest_file = f"data/output_{i}.mp3"
        try:
            shutil.copy(OUTPUT_FILE, dest_file)
        except Exception as e:  print("Error on copy:", str(e))
if __name__ == "__main__":
    start_num = 90
    end_num = 120
    main()
    try:
        display(Audio(f"data/test.mp3", autoplay=True))
    except Exception as e:  print("Error:", str(e))
```
